```python
# Find the city having the highest average sales.
a= df.groupby('city')['sales'].mean()

for city, price in a.items():
    if price == a.max():
        print(f"City name {city} Average maximun value is {price}")
```

```python
# Find the category contributing the highest percentage of total sales.

m = df.groupby('category')['sales'].sum() / df['sales'].sum()*100
# for c, i in m.items():
#     if i == m.max():
#         print(c, i )


print(m.idxmax(), m.max())
```

```python
# Find states where total sales exceed 50,000.
state_total= df.groupby('state')['sales'].sum()
state_total[state_total>50000]
```

```python
# Find the top customer from each region based on sales.
group_region= df.groupby(['region','customer_name'])['sales'].sum().reset_index()
top_cus= group_region.loc[group_region.groupby('region')['sales'].idxmax()]

top_cus
```

```python
# Find the top customer from each region based on sales.
reg= df.groupby(['region','customer_name'])['sales'].sum().reset_index()

reg['rank']=reg.groupby('region')['sales'].rank(method='dense',ascending=False)
reg[reg['rank']==1]
```
